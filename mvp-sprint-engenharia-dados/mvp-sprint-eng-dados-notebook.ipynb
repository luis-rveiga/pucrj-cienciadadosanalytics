**Produto Interno Bruto per capita** - https://dados.gov.br/dados/conjuntos-dados/cgeo_vw_pib_percapita

O Produto Interno Bruto - PIB per capita de cada município foi estimado pelo quociente entre o valor do PIB do municípioe sua população residente. Para a população, foi utilizada a estimativa encaminhada pelo IBGE ao Tribunal de Contas da União- TCU, em outubro de 2009, tendo 1º de julho de 2009 como data de referência.Produto interno bruto é o total dos bens e serviços produzidos pelas unidades produtoras residentes destinados ao consumofinal sendo, portanto, equivalente à soma dos valores adicionados pelas diversas atividades econômicas acrescida dos impostos,líquidos de subsídios, sobre produtos. O produto interno bruto também é equivalente à soma dos consumos finais de bense serviços valorados a preço de mercado sendo, também, equivalente à soma das rendas primárias.

```python
bolsa_familia_2014_path = bucket_path + "bolsa_familia_2014.txt"
bolsa_familia_2015_path = bucket_path + "bolsa_familia_2015.txt"
bolsa_familia_2016_path = bucket_path + "bolsa_familia_2016.txt"
bolsa_familia_2017_path = bucket_path + "bolsa_familia_2017.txt"
bolsa_familia_2018_path = bucket_path + "bolsa_familia_2018.txt"
bolsa_familia_2019_path = bucket_path + "bolsa_familia_2019.txt"
```

```python
bolsa_familia_2014_2019_df = spark.read.option("inferSchema",True).option("header",True).csv(bolsa_familia_2014_path)
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.union(spark.read.option("inferSchema",True).option("header",True).csv(bolsa_familia_2015_path))
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.union(spark.read.option("inferSchema",True).option("header",True).csv(bolsa_familia_2016_path))
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.union(spark.read.option("inferSchema",True).option("header",True).csv(bolsa_familia_2017_path))
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.union(spark.read.option("inferSchema",True).option("header",True).csv(bolsa_familia_2018_path))
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.union(spark.read.option("inferSchema",True).option("header",True).csv(bolsa_familia_2019_path))
```

```python
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.withColumn("valor_medio_beneficio", bolsa_familia_2014_2019_df.valor_repassado_bolsa_familia / bolsa_familia_2014_2019_df.qtd_familias_beneficiarias_bolsa_familia)
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.withColumn("ano", bolsa_familia_2014_2019_df.anomes.substr( 1, 4).cast("int"))
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.withColumn("mes", bolsa_familia_2014_2019_df.anomes.substr( 5, 2).cast("int"))
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.withColumnRenamed("ibge","codigo_regiao")
bolsa_familia_2014_2019_df = bolsa_familia_2014_2019_df.drop("anomes")
```

```python
spark.sql("DROP TABLE IF EXISTS bolsa_familia_2014_2019_table")
bolsa_familia_2014_2019_df.write.saveAsTable("bolsa_familia_2014_2019_table")
```

```python
bolsa_familia_2014_2019_df = spark.sql('''
    select codigo_regiao, ano, sum(qtd_familias_beneficiarias_bolsa_familia)/count(qtd_familias_beneficiarias_bolsa_familia) media_qtd_familias_beneficiarias_bolsa_familia, sum(valor_repassado_bolsa_familia)/count(valor_repassado_bolsa_familia) media_valor_repassado_bolsa_familia, sum(valor_medio_beneficio)/count(valor_medio_beneficio) media_valor_medio_beneficio from bolsa_familia_2014_2019_table group by codigo_regiao, ano
    ''')
```
